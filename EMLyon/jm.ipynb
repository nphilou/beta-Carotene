```python
grouped = df.groupby('location')
grouped.groups
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.
# Create a Pandas dataframe from the data.
emptydf = pd.DataFrame({})

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('pandas_simple.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
emptydf.to_excel(writer, sheet_name='Sheet1')

# Get the xlsxwriter objects from the dataframe writer object.
workbook  = writer.book
worksheet1 = writer.sheets['Sheet1']

xrow = 0
for name, group in grouped:
    startrow = xrow
    for index, row in group.reset_index(drop=True).iterrows():
        worksheet1.set_row(xrow, None, None, {'level': 2})
        xrow += 1
    group.to_excel(writer, sheet_name='Sheet1', startrow=startrow, header=False, index=False)
        
    worksheet1.set_row(xrow, None, None, {'level': 1})
    worksheet1.write(xrow, 1, name)
    worksheet1.write(xrow, 2, '=SUBTOTAL(9,C%s:C%s)' % (startrow + 1, xrow))
    xrow += 1

workbook.close()
```
